## Valid Vectors for classification 


We remove all those cells were it is not possible to develop new urban cells like, the existing urbanized cells, industry, water shapes, protected areas, roads, rails. 

```python


validVectors = suelo[~(
    (suelo.urbano_2000c_full == 1) | 
    (suelo.industrias_full == 1) | 
    (suelo.aguac_full == 1)| 
    (suelo.areas_naturales_full == 1) | 
    (suelo.vialidades_full == 1) | 
    (suelo.via_ferrea_full == 1)
) ]


fullSample = validVectors[{u'lat',"long", u'd_urbano_2000c', u'd_Industrias',u'd_Aguac',u'd_Areas_naturales', u'd_Vialidades',u'd_puntos',u'd_via_ferrea',u'road_dens_perc', u'tiempo_viaje_ciudad', u'pendiente',u'urbano_2010c_full' }]

#we save the valid classification vectors in a new csv.
fullSample.to_csv("muestrasEntrenamiento.csv")
```
